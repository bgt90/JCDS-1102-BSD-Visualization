# Matplotlib Overview

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#buat data dummy
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
y
```

matplotlib punya 3 layer, yaitu:
* Layer Backend
* Layer Artist __> jika ingin memoadifikasi lebih, pakai layer ini
* Layer scripting --> user biasanya memakai layer ini

## Basic Matplotlib: Scripting Layer

```python
plt.figure(figsize=(12,7))
plt.plot(x,y,'r') #plot x di axis x, y di axis y, beri warna merah
plt.xlabel('ini x', size=15)
plt.xticks([1,2,3,4,5],['Andi','Budi','Caca','Dedi','Edi']) #andi, budi,caca, dedi dan edi di posisi 1-5
plt.ylabel('ini y',size=15)
plt.yticks([1,2,3,4,5])
plt.title('ini judul',size=20)
plt.show()
```

```python
#2 plot dalam 1 grafik
plt.plot(x,y,'r--')
plt.plot(y,x,'g*')
plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.subplot(1,2,1) #baris kolom posisi
plt.plot(x,y,'r--')

plt.subplot(122)
plt.plot(y,x,'g*')
```

```python
plt.figure(figsize=(12,8))
plt.subplot(221)
plt.plot(x,y,'r--')
plt.xlabel('ini x1')
plt.ylabel('ini y1')
plt.title('ini judul 1')

plt.subplot(222)
plt.plot(y,x,'g*')
plt.xlabel('ini x2')
plt.ylabel('ini y2')
plt.title('ini judul 2')

plt.subplot(223)
plt.plot(x,y,'b')
plt.xlabel('ini x3')
plt.ylabel('ini y3')
plt.title('ini judul 3')

plt.suptitle('ini judul besar')
plt.show()
```

## Artist Layer

```python
#buat kanvas/figure kosong
fig=plt.figure(figsize=(10,10))

#tambah axis
axes=fig.add_axes([0.1,0.1,0.5,0.5]) #posisi axis (kiri, bawah, lebar, tinggi (angka float 0-1))

# buat plot yang diinginkan
axes.plot(x,y,'b--')
axes.set_xlabel('ini sumbu x')
axes.set_ylabel ('ini sumbu y')
axes.set_title('ini judul')
plt.show()
```

```python
fig=plt.figure(facecolor='yellow')

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b*')
axes1.set_xlabel('ini sumbu x1')
axes1.set_ylabel('ini sumbu y1')
axes1.set_title('ini judul1')

axes2.plot(y,x,'r--')
axes2.set_xlabel('ini sumbu x2')
axes2.set_ylabel('ini sumbu y2')
axes2.set_title('ini judul2')
# axes2.set_axis_off()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(8,8),facecolor='yellow')

for ax in axes:
    # ax.plot(x,y,'b')
    ax.set_xlabel('ini x')
    ax.set_ylabel('ini y')
    ax.set_title('ini judul')
# axes # berisi array posisi subplot

axes[1].plot(x,y)
axes[0].plot(y,x,'r--')

```

```python
fig,axes =plt.subplots(nrows=2,ncols=2,figsize=(8,8),facecolor='red')

# axes[0]

# for i in axes:
    # for ax in i:
    #     ax.set_xlabel('ini x')
    #     ax.set_ylabel('ini y')
    #     ax.set_title('ini judul')

axes[0][0].plot(x,y,'r--',label='ini 1')
axes[0][0].plot(y,x,'g*', label='ini 2')
axes[0][0].legend() #kalau mau pakai legend, maka plot di axis perlu dilabel 
# plt.tight_layout() #ukuran grafik memenuhi kanvas
```

# Line plotting stock dataset

```python
#line plot sering dipakai untuk time series
```

```python
cleo= pd.read_csv('CLEO.JK.csv')
icbp= pd.read_csv('ICBP.JK.csv')
myor=pd.read_csv('MYOR.JK.csv')
ultj=pd.read_csv('ULTJ.JK.csv')
```

```python
cleo.head()
```

```python
cleo.isna().sum()
```

```python
cleo.dtypes #cek datatype di tiap kolom
```

```python
type(cleo['Date'][0]) 
```

```python
cleo['Date']=pd.to_datetime(cleo['Date']) #ubah tipe data "Date" jadi tipe data tanggal
```

```python
cleo.info()
```

```python
## lineplot
```

```python
x=cleo['Date']
y=cleo['Open']

plt.figure(figsize=(10,6))
plt.plot(x,y,'r--')
plt.xlabel('Tanggal')
plt.ylabel('Harga Pembuka Saham')
plt.title('Harga Buka Harian Saham CLEO')
```

```python
x=cleo['Date']
y1=cleo['Open']
y2=ultj['Open']

plt.figure(figsize=(10,6))
plt.plot(x,y1,'r--',x,y2,'b')
plt.xlabel('Tanggal')
plt.ylabel('Harga Pembuka Saham')
plt.title('Harga Buka Harian Saham CLEO dan ULTJ', size=20)
plt.legend(['Cleo','Ultrajaya'],loc=0)
plt.show()
```

```python
x=cleo['Date']
y1=cleo['Open']
y2=ultj['Open']
y3=myor['Open']
y4=icbp['Open']

plt.figure(figsize=(10,6))
plt.plot(x,y1,'r--',x,y2,'b',x,y3,'go',x,y4,'yellow')
plt.xlabel('Tanggal')
plt.xticks(rotation=45)
plt.ylabel('Harga Pembuka Saham')
plt.suptitle('Histori Harga Saham Perusahaan Makanan dan Minuman Tahun 2020', size=20)
plt.legend(['CLEO','ULTJ','MYOR','ICBP'],bbox_to_anchor=(0.,1,1.,.102),
    loc='lower right', ncol=4, mode='expand',borderaxespad=0)
plt.show()
```

```python
x=cleo['Date']
y1=cleo['Open']
y2=ultj['Open']
y3=myor['Open']
y4=icbp['Open']

plt.style.use('ggplot')
fig=plt.figure(figsize=(15,7))

ax1=fig.add_subplot(221)
ax1.plot(x,y1,'orange')
ax1.set_title('CLEO')

ax2=fig.add_subplot(222)
ax2.plot(x,y2,'b')
ax2.set_title('ULTJ')

ax3=fig.add_subplot(223)
ax3.plot(x,y3,'darkblue')
ax3.set_title('MYOR')

ax4=fig.add_subplot(224)
ax4.plot(x,y4,'hotpink')
ax4.set_title('ICBP')

fig.suptitle('Histori Harga Saham Makanan dan Minuman 2020',size=20)
plt.show()
```

## Stackplot

```python
x=cleo['Date']


y=np.vstack([cleo['Open'],myor['Open'],ultj['Open'], icbp['Open']]) #menggabungkan data secara vertikal
```

```python
y
```

```python
plt.figure(figsize=(15,7))

labels=['CLEO','MYOR','ULTJ','ICBP']
color=['hotpink','royalblue','gold','lime']
plt.stackplot(x,y,labels=labels,colors=color,edgecolor='k') #edgecolor:memberi warna border pada stack
plt.legend(loc=0)
plt.show()
```

```python
### percentange stacked area
```

```python
#stackplot with percentage data
x=cleo['Date']
y1=cleo['Open'].divide(cleo['Open'].sum())
y2=myor['Open'].divide(myor['Open'].sum())
y3=ultj['Open'].divide(ultj['Open'].sum())
y4=icbp['Open'].divide(icbp['Open'].sum())
labels=['CLEO','MYOR','ULTJ','ICBP']
color=['hotpink','royalblue','gold','lime']

plt.figure(figsize=(10,7))
plt.stackplot(x,y1,y2,y3,y4,labels=labels,colors=color,edgecolor='k')
plt.legend(loc=0)
plt.show()
```

```python
### stemplot
```

```python
x=cleo['Date']
y=cleo['Open'].diff() #selisih antara data di dalam kolom
```

```python
plt.figure(figsize=(10,5))
plt.stem(x,y,markerfmt='b_',linefmt='r--',basefmt='y:', use_line_collection=True)
plt.title('Pergerakan Harian Harga Saham Cleo 2020')
plt.show()
```

```python
x=cleo['Date']
y1=cleo['Open'].diff()
y2=myor['Open'].diff()
y3=ultj['Open'].diff()
y4=icbp['Open'].diff()

plt.figure(figsize=(12,20))

plt.subplot(411)
plt.stem(x,y1,markerfmt='k_',linefmt='r--',basefmt='g:', use_line_collection=True)
plt.title('CLEO',size=15)

plt.subplot(412)
plt.stem(x,y2,markerfmt='k_',linefmt='b--',basefmt='g:', use_line_collection=True)
plt.title('MYOR',size=15)

plt.subplot(413)
plt.stem(x,y3,markerfmt='k_',linefmt='y--',basefmt='g:', use_line_collection=True)
plt.title('ULTJ',size=15)

plt.subplot(414)
plt.stem(x,y4,markerfmt='k_',linefmt='m--',basefmt='g:', use_line_collection=True)
plt.title('ICBP',size=15)

# plt.suptitle('Perubahan Pembukaan Harga Saham Harian 2020',size=18)
plt.show()
# plt.tight_layout()
```
